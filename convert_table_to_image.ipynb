## Brief-Dateien aktualisieren

```python
from bs4 import BeautifulSoup

import glob, os
from pathlib import Path


import glob
# es werden nur die Brief-Dateien aktualisiert
files = glob.glob('./result/OEBPS/Text/Brief*.xhtml')



for file in files:
    
   # Dateiname filtern
    name = os.path.basename(file).split('.xhtml')[0]

    xml_file = open(file, encoding = 'utf-8')
    data = xml_file.read()     
    #html-Dateien laden 
    soup = BeautifulSoup(data,'xml')
    speak = soup.find_all('table')
   
    hti = Html2Image(output_path=f'./result/OEBPS/Images/')
    css = "table {background-color:white;height:100%;width:100%; position: absolute; top: 0; bottom: 0; left: 0; right: 0;border:1px solid;table-layout:auto;} table,td,th { border-collapse: collapse;}th,td {padding: 10px;border: solid 1px;text-align: center;}th{font:size:0.3em;}td{font-size:0.7em;}"
    paths = hti.screenshot(html_str=speak, css_str=css,save_as=f'{name}_table.png')
    
    
    
    for idx, sp in enumerate(speak):
       
            
    
      # neuen tag erstellen
        new_cap = soup.new_tag("figcaption")
        new_cap.string=f"{name}_Tabelle_{idx}"
        
        
        new_tag = soup.new_tag("figure")
        new_tag['class'] = 'image'
        
       
        
        tag = soup.new_tag("img")
        # als Attribut wird der Pfad zum neuen Bild geschrieben
        tag['src'] = f'../Images/{name}_table_{idx}.png'
        tag['class'] = 'fallback_table'
        
        
        new_tag.append(tag)
        new_tag.append(new_cap)
    # nach jeder Tabelle wird ein Bild der Tabelle eingefügt
        
        sp.insert_after(new_tag)
    # die neuen Dateien werden gespeichert
    with open(file, "w", encoding='utf-8') as f:
        f.write(str(soup))
```
